```python
print(1==1)
print(1<2)
print(1>2)
print(2>=2)
print(2<=2)
print(2!=2) #다르다
```

```python
#if 조건문(불자료형): 수행할 문장들
print(bool("python"))
print(bool(""))#이거를 제외하면 다 트루(문자일경우)
print(bool(" "))
print(bool(10))
print(bool(0))#0말고는 다 트루(숫자일경우)
print(bool([1,2,3]))
print(bool([]))#리스트일 경우 아무것도 없으면 false
print(bool({"name":"pey"}))
print(bool({}))#아무것도 없으면 false

```

```python
#논리연산자 
##and 연산자 # 다 True일때만 True가 나온다
print(True and True )
print(True and False )
print(False and True )
print(False and False )
print(True and True and False)
```

```python
#or연산자 
print(True or True )
print(True or False )
print(False or True )
print(False or False )
print(True or True or False)
```

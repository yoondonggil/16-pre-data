```python
treehit = 0
while treehit <10:
    treehit = treehit +1
    print(f'나무를{treehit}번 찍었습니다')
    if treehit == 10:
        print('나무가 넘어갑니다')
```

```python
coffee = 5
money = 300
while money:
    print('돈을 받았으니 커피를 줍니다.')
    coffee = coffee -1
    print(f'남은 커피는 {coffee}개 입니다')
    
    if coffee  == 0:
        print('커피가 다 떨어졌습니다')
        break
```

```python
coffee = 5 
while True:
    money =int(input('돈을 넣어 주세요:'))
    if money == 300:
        print('커피를 줍니다')
        coffee -= 1 #coffee=coffee -1 과 같다
    elif money>300:
        print(f'거스름돈 {-300}를 주고 커피를 줍니다')
        coffee -=1
    else:
        print('돈을 다시 돌려주고 커피를 주지 않습니다')
        print(f'현재 남은 커피는{coffee}개 입니다')
    if coffee==0:
        print('커피가 다 떨어졌습니다')
        break
```

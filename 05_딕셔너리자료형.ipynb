```python
#딕셔너리는 어떻게 만들까?
#{key1: value1,key2:value2,...}


dic = {
    'name':'pey',
    'phone':'01012341234',
    'birth':'1118'
}
print(dic)
print(type(dic))
```

```python
#딕셔너리 쌍추가하기
a={1:"a"}
a[2]='b'  #[]는 key값이다 앞에서 했던 index가 아니고 key 값을 설정하는것이다
print(a)
```

```python
a[3] = [1,2,3]
print(a)
```

```python
#딕셔너리 요소 삭제하기
print(a)
del a[1]
print(a)
```

```python
#딕셔너리에 key를 사용해서 value 얻기 
grade={
    'pey':10,
    'juli':99
}
print(grade['pey'])
print(grade['juli'])
```

```python
#딕셔너리 관련 함수
## 1. key리스트 만들기-keys

a={
    'name':'pey',
    'phone':'01012341234',
    'birth':'1118'
}
a.keys()
list(a.keys())
```

```python
#value 리스트 만들기 -values
a.values()
list(a.values())
```

```python
## key , value 쌍얻기 -items 
print(a)
a.items()
list(a.items())
```

```python
##key,value 쌍 모두 지우기=clear
a.clear()
print(a)
```

```python
#key로 value 얻기-get 
a= {
    'name':'pey',
    'phone':'01012341234',
    'birth':'1118'
}
print(a['name'])
print(a.get('name'))
```

```python
print(a['hobby']) #없으니까 error 뜸
a.get('hobby') # 없지만 error는 안뜸 그냥 안 나옴
a.get('hobby',0)
```

```python
#해당 key가 딕셔너리 안에 있는지 조사하기-in
print(a)
print('name' in a)
print('email' in a)
```

```python
# key로 value 꺼내기 -pop-->value값만 나옴 
a= {
    'name':'pey',
    'phone':'01012341234',
    'birth':'1118'
}
result=a.pop('phone')
print(result)
print(a)
```

```python
#딕셔너리의 key값으로 올 수 없는 데이터 타입은 list-mutable

a=[1,2,3]
a[0]=3
print(a)
```

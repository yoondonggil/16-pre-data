```python
marks =[90,25,67,45,80]
number = 0
for mark in marks:
    number=number+1
    if mark>=60:
        print(f'{number}학생은 합격입니다')
    else:
        print(f'{number}학생은 불합격입니다')
        
del number
```

```python
# for문 continue
marks=[90,25,67,45,80]
number =0
for mark in marks:
    number+=1
    if mark<60:
        continue
print(f'{number}학생은 합격입니다')
```
